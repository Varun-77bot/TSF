```python

import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
```

```python
plt.figure(figsize=(15,6))
plt.scatter(data['Hours'],data['Scores'])
plt.title('Hours Vs Percentage')
plt.xlabel('Hours')
plt.ylabel('Percentage of Scores')
```
